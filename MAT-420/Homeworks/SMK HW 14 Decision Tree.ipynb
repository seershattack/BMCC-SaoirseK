```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree, DecisionTreeRegressor
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
dogs_train = pd.read_csv('dogs_train.csv')
X_train = dogs_train.drop(['BreedGroupId', 'Breed', 'Age segment'], axis = 1)
y_train = dogs_train['BreedGroupId']
```

```python
model = DecisionTreeClassifier(criterion = "entropy") # Just to see the chaos
model.fit(X_train, y_train)

plt.figure(figsize = (12, 8))
plot_tree(model, filled = True, rounded = True)

plt.show()
```

```python
tree_classifier2 = DecisionTreeClassifier(criterion = "entropy", min_impurity_decrease = 0.001, max_depth = 2)
tree_classifier2.fit(X_train, y_train)

plt.figure(figsize=(12, 8))
plot_tree(tree_classifier2, filled=True, rounded=True)

plt.show()
```

```python
pred = model.predict(X_train)
confusion_matrix(y_train, pred)
```
